Inertia: 
It relies on the concept of inertia, which is defined as follows: the sum of the Euclidean distances between each point and its associated centroid.

In data management, Inertia measures the spread or variance of data points around the mean

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

np.random.seed(0)  
cluster_1 = np.random.rand(50, 2) + [2, 2]   # 50 points around (2, 2)
cluster_2 = np.random.rand(50, 2) + [8, 8]   # 50 points around (8, 8)
data = np.concatenate((cluster_1, cluster_2), axis=0)



# 2. Initialize and fit the K-Means model
# We know we generated 2 clusters, so let's set k=2
kmeans = KMeans(n_clusters=2, n_init=10) # n_init to avoid local optima
kmeans.fit(data)

# 3. Get the cluster assignments and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# 4. Visualize the clusters and centroids
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', label='Data Points')  ## c for color, cmap refers colormap
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')
plt.title('Simple K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 5. Print the cluster assignments and centroids (optional)
print("Cluster Assignments:")
print(labels)
print("\nCentroids:")
print(centroids)
```
