```python
import plotly.graph_objects as go
import numpy as np

species = ("1990", "2000", "2010")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

# Create a figure
fig = go.Figure()

# Add bars for each attribute
for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    fig.add_trace(go.Bar(
        x=x + offset, 
        y=measurement, 
        width=width,
        name=attribute
    ))
    multiplier += 1

# Update layout
fig.update_layout(
    title="Penguin attributes by species",
    xaxis=dict(
        tickmode='array',
        tickvals=x + width,
        ticktext=species,
    ),
    yaxis=dict(
        title="Length (mm)",
        range=[0, 250],
    ),
    barmode='group',
    legend=dict(
        title="Attributes",
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5,
    ),
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Time intervals (years)
years = ['1990', '2000', '2010', '2020']
countries = ['Sweden', 'Norway', 'Denmark']
results = ['win', 'lost', 'draw']
colors = {
    'win': '#2ca02c',   # Muted teal for Win
    'lost': '#ff7f0e',  # Muted red/rose for Lost
    'draw': '#1f77b4'   # Soft blue for Draw
}


# Win data
data = {
    'Sweden': {
        '1990': {'win': 2, 'lost': 4, 'draw': 4},
        '2000': {'win': 6, 'lost': 2, 'draw': 2},
        '2010': {'win': 4, 'lost': 4, 'draw': 0},
        '2020': {'win': 2, 'lost': 1, 'draw': 0},
    },
    'Norway': {
        '1990': {'win': 4, 'lost': 2, 'draw': 4},
        '2000': {'win': 2, 'lost': 4, 'draw': 2},
        '2010': {'win': 8, 'lost': 1, 'draw': 2},
        '2020': {'win': 2, 'lost': 2, 'draw': 2},
    },
    'Denmark': {
        '1990': {'win': 0, 'lost': 4, 'draw': 2},
        '2000': {'win': 1, 'lost': 2, 'draw': 2},
        '2010': {'win': 1, 'lost': 2, 'draw': 4},
        '2020': {'win': 8, 'lost': 1, 'draw': 1},
    }
}

# Create figure
fig = go.Figure()
x = np.arange(len(years))  # positions for years
width = 0.2  # bar width
offsets = [-width, 0, width]  # group offsets for 3 countries

# Add bars for each country and result type
for i, country in enumerate(countries):
    wins = []
    losts = []
    draws = []

    for year in years:
        wins.append(data[country][year]['win'])
        losts.append(data[country][year]['lost'])
        draws.append(data[country][year]['draw'])

    x_offset = x + offsets[i]

    # Only show legend for the first country
    show_legend = (i == 0)

    # Win
    fig.add_trace(go.Bar(
        x=x_offset,
        y=wins,
        name='Win',
        marker_color=colors['win'],
        width=width,
        showlegend=show_legend
    ))

    # Lost
    fig.add_trace(go.Bar(
        x=x_offset,
        y=losts,
        name='Lost',
        marker_color=colors['lost'],
        width=width,
        base=wins,
        showlegend=show_legend
    ))

    # Draw
    total_base = np.array(wins) + np.array(losts)
    fig.add_trace(go.Bar(
        x=x_offset,
        y=draws,
        name='Draw',
        marker_color=colors['draw'],
        width=width,
        base=total_base,
        showlegend=show_legend
    ))

# Update layout
fig.update_layout(
    title="Country-wise Win/Lost/Draw per Year (Grouped + Stacked)",
    xaxis=dict(
        tickmode='array',
        tickvals=x,
        ticktext=years,
        title='Year'
    ),
    yaxis=dict(
        title='Count',
        range=[0, 15]
    ),
    barmode='stack',
    legend=dict(
        title='Match Result',
        orientation='h',
        yanchor='bottom',
        y=1.02,
        xanchor='center',
        x=0.5
    )
)

fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Create dummy data indexed by state and with multi-columns [product, revenue]
index = ["1971-1980", "1981-1990", "1991-2000", "2001-2010", "2011-2020"]

df = pd.concat(
    [
        pd.DataFrame(
            np.random.rand(5, 3) * 1.25 + 0.25,
            index=index,
            columns=["Revenue1", "Revenue2", "Revenue3"]
        ),
        pd.DataFrame(
            np.random.rand(5, 3) + 0.5,
            index=index,
            columns=["Revenue1", "Revenue2", "Revenue3"]
        ),
    ],
    axis=1,
    keys=["Product1", "Product2"]
)

# Calculate y-axis range without using deprecated groupby(axis=1)
yaxis_max = df.T.groupby(level=0).sum().T.max().max() * 1.5

# Create the figure with the updated layout
fig = go.Figure(
    layout=go.Layout(
        height=600,
        width=1000,
        barmode="relative",
        yaxis_showticklabels=False,
        yaxis_showgrid=False,
        yaxis_range=[0, yaxis_max],
        # Secondary y-axis overlayed on the primary one and not visible
        yaxis2=go.layout.YAxis(
            visible=False,
            matches="y",
            overlaying="y",
            anchor="x",
        ),
        font=dict(size=24),
        legend_x=0,
        legend_y=1,
        legend_orientation="h",
        hovermode="x",
        margin=dict(b=0, t=10, l=0, r=10)
    )
)

# Define some colors for the product, revenue pairs
colors = {
    "Product1": {
        "Revenue1": "#F28F1D",
        "Revenue2": "#F6C619",
        "Revenue3": "#FADD75",
    },
    "Product2": {
        "Revenue1": "#2B6045",
        "Revenue2": "#5EB88A",
        "Revenue3": "#9ED4B9",
    }
}

# Add the traces
for i, product in enumerate(colors):
    for j, revenue in enumerate(df[product].columns):
        if (df[product][revenue] == 0).all():
            continue
        fig.add_bar(
            x=df.index,
            y=df[product][revenue],
            yaxis=f"y{i + 1}",
            offsetgroup=str(i),
            offset=(i - 1) * 1/3,
            width=1/3,
            legendgroup=product,
            legendgrouptitle_text=product,
            name=revenue,
            marker_color=colors[product][revenue],
            marker_line=dict(width=2, color="#333"),
            hovertemplate="%{y}<extra></extra>"
        )

# Show the figure
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Data
years = [2020, 2021, 2022]
acases = [180, 190, 200]
aclosed = [90, 90, 90]
bcases = [220, 200, 220]
bclosed = [165, 150, 165]
ccases = [90, 110, 100]
cclosed = [70, 100, 90]

# Create DataFrame
df = pd.DataFrame({
    'Year': years,
    'Cases_A': acases,
    'Closed_A': aclosed,
    'Cases_B': bcases,
    'Closed_B': bclosed,
    'Cases_C': ccases,
    'Closed_C': cclosed,
})

# Add open cases
df['Open_A'] = df['Cases_A'] - df['Closed_A']
df['Open_B'] = df['Cases_B'] - df['Closed_B']
df['Open_C'] = df['Cases_C'] - df['Closed_C']

# Define colors
colors = {'Closed': '#018571', 'Not Closed': '#a6611a'}

# Units
units = ['A', 'B', 'C']

# Build x-axis labels and bar traces
fig = go.Figure()
x_labels = []
closed_y = []
open_y = []
unit_labels = []

for i, year in enumerate(df['Year']):
    for unit in units:
        label = f"{year} {unit}"
        x_labels.append(label)
        closed_y.append(df[f'Closed_{unit}'][i])
        open_y.append(df[f'Open_{unit}'][i])
        unit_labels.append(unit)

# Now create stacked bars per x position
for idx, status in enumerate(['Closed', 'Not Closed']):
    for unit in units:
        y_vals = [df[f'{status if status == "Closed" else "Open"}_{unit}'][i] for i in range(len(df))]
        x_vals = [f"{df['Year'][i]} {unit}" for i in range(len(df))]
        fig.add_trace(go.Bar(
            x=x_vals,
            y=y_vals,
            name=status,
            marker_color=colors[status],
            offsetgroup=unit,
            base=None if status == 'Closed' else [df[f'Closed_{unit}'][i] for i in range(len(df))],
            legendgroup=status,
            showlegend=(unit == 'A')  # show legend only once per status
        ))

# Layout
fig.update_layout(
    barmode='stack',
    title='Cases Assigned by Unit 21–23',
    xaxis_title='Year and Unit',
    yaxis_title='Number of Cases',
    legend_title='Status',
    bargap=0.3
)

fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = [2020, 2021, 2022]
acases = [180, 190, 200]
aclosed = [90, 90, 90]
bcases = [220, 200, 220]
bclosed = [165, 150, 165]
ccases = [90, 110, 100]
cclosed = [70, 100, 90]

# Compute open cases
aopen = [a - c for a, c in zip(acases, aclosed)]
bopen = [b - c for b, c in zip(bcases, bclosed)]
copen = [c - cl for c, cl in zip(ccases, cclosed)]

# Bar width and positions
bar_width = 0.2
x = np.arange(len(years))  # base x locations per year

# Offsets for each unit
offsets = [-bar_width, 0, bar_width]  # A, B, C
units = ['A', 'B', 'C']
closed = [aclosed, bclosed, cclosed]
open_ = [aopen, bopen, copen]

colors = {'Closed': '#018571', 'Not Closed': '#a6611a'}

fig, ax = plt.subplots(figsize=(10, 6))

for i in range(3):  # For units A, B, C
    x_pos = x + offsets[i]
    ax.bar(x_pos, closed[i], width=bar_width, color=colors['Closed'], label='Closed' if i == 0 else "", edgecolor='black')
    ax.bar(x_pos, open_[i], width=bar_width, bottom=closed[i], color=colors['Not Closed'], label='Not Closed' if i == 0 else "", edgecolor='black')

# X-axis and legend
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.set_xlabel('Year')
ax.set_ylabel('Number of Cases')
ax.set_title('Cases Assigned by Unit 21–23')
ax.legend()

plt.tight_layout()
plt.show()
```
