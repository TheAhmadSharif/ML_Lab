```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
np.random.seed(12)
k = 2
```

```python
cluster_1 = np.array([[2.0, 2.0],
                      [2.0, 2.2],
                      [8.0, 2.0],
                      [8.0, 2.2]])

cluster_2 = np.array([[8.0,  8.0],
                      [8.0,  8.2],
                      [2.0, 8.0],
                      [2.0, 8.2]])
data = np.concatenate((cluster_1, cluster_2), axis=0)
kmeans = KMeans(n_clusters = k, random_state=12, n_init = 10)
kmeans.fit(data)
get_labels = kmeans.labels_
get_centriods = kmeans.cluster_centers_



# https://se.mathworks.com/help/matlab/ref/colormap.html

cmap_list = ['summer', 'winter', 'spring', 'autumn']

plt.figure(figsize=(10, 8))
plt.scatter(data[:, 0], data[:, 1], c = get_labels,  cmap='winter', label='Data Points')
plt.scatter(get_centriods[:, 0], get_centriods[:, 1], marker='X', s=200, color='red', label='Centroids')



# cmap : summer, winter, spring, autumn

for i, centroid in enumerate(get_centriods):
    plt.annotate(f'Centroid {i+1}', xy=(centroid[0], centroid[1]), xytext=(centroid[0]+0.1, centroid[1]+0.1))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D  # Import for 3D plotting

# Define the data with three factors
cluster_1 = np.array([[2.0, 2.0, 1.0],
                      [2.0, 2.2, 1.5],
                      [2.2, 2.0, 2.0],
                      [2.2, 2.2, 2.5]])

cluster_2 = np.array([[8.0, 8.0, 5.0],
                      [8.0, 8.2, 5.5],
                      [8.2, 8.0, 6.0],
                      [8.2, 8.2, 6.5]])

data = np.concatenate((cluster_1, cluster_2), axis=0)
k = 2

# Perform K-Means clustering
kmeans = KMeans(n_clusters=k, random_state=12, n_init=10)
kmeans.fit(data)
get_labels = kmeans.labels_
get_centriods = kmeans.cluster_centers_

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')  # 111 means 1x1 grid, first subplot
ax.scatter(data[:, 2], data[:, 1], data[:, 0], c=get_labels, cmap='winter', label='Data Points')
ax.scatter(get_centriods[:, 0], get_centriods[:, 1], get_centriods[:, 2], marker='X', s=200, color='red', label='Centroids')

# Labeling the centroids
for i, centroid in enumerate(get_centriods):
    ax.text(centroid[0], centroid[1], centroid[2], f'Centroid {i+1}', color='black')

ax.set_xlabel('Factor 1')
ax.set_ylabel('Factor 2')
ax.set_zlabel('Factor 3')
plt.title('K-Means Clustering with 3-Factor Data')
plt.legend()
plt.show()

print(data)
```
