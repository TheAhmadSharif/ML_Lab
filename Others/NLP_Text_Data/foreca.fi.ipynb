```python
import requests
import bs4
import nltk
import operator
import string
import pprint

from nltk.corpus import stopwords
web_url = 'https://www.foreca.fi/'
web_html = requests.get(web_url)


clean_text = ' '.join(bs4.BeautifulSoup(web_html.content, "html.parser").stripped_strings)

res = nltk.word_tokenize(clean_text)
len(res)
```

```python
token_list = nltk.word_tokenize(clean_text)
Text = clean_text
stop = set(stopwords.words('finnish') + list(string.punctuation))
token_list = nltk.word_tokenize(Text)
filtered_token = [w for w in token_list if not w in stop]


freq_word = {}

for word in filtered_token:
    if(word in freq_word):
        freq_word[word] += 1
    else:
        freq_word[word] = 1
        

word_limit = 500
        
sorted_freq_dist = sorted(freq_word.items(), key = operator.itemgetter(1), reverse=True)[:word_limit]



# Prints the nicely formatted dictionary
pprint.pprint(sorted_freq_dist)

```
