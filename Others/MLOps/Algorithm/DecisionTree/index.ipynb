```python
""" 
28 March 2023
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from math import log2, log
dataset = data = pd.read_csv("weather.csv")


print(data)

print()
play_yes =  sum((data.play == "yes"))
play_no =  sum((data.play == "no"))
total_game = len(data)
total_game

print("Total Play    ", play_yes)
print("Total Not Play", play_no)

play_yes_proportion = play_yes/total_game
play_no_proportion = play_no/total_game

Entrophy = (-(play_yes_proportion * log2(play_yes_proportion)) - (play_no_proportion * log2(play_no_proportion)))
Entrophy
```

```python
dataset = data = pd.read_csv("weather.csv")
# data[data["outlook"] == "overcast"]
#data[["outlook", "play", "temperature"]]

# E_Outlook = (4/14)*E(4,0) + (4/14)*log2(4/14) + (4/14)*log2(4/14) 
data1 = data[data["outlook"] == "sunny"][["outlook", "play"]].sort_values(by=['play'])
print(data1, '_______ hot _______')
print()

data2 = data[data["outlook"] == "rainy"][["outlook", "play"]].sort_values(by=['play'])
print(data2, '_______ cool _______')
print()



# E_Outlook = 
# (2/14)*E(2,3) # sunny 
# (3/14)*E(3,2) # Rainy
# (4/14)*E(4,0) # overcast
E_Outlook = (4/14)*((-4/4)*log2(4/14)) + (5/14)*((-2/5)*log2(2/5) - (3/5)*log2(3/5)) + (5/14)*((-3/5)*log2(3/5) -(2/5)*log2(2/5))
print(E_Outlook)

print()
IG_S_Outlook = Entrophy - E_Outlook
print(IG_S_Outlook, '___ IG_S_Outlook ___') # -0.2696373

## Entropy Temperature ## 

print('______________')
print()
E_Temperature = (4/14)*((-2/4)*log2(2/4) -(2/4)*log2(2/4)) + (4/14)*((-3/4)*log2(3/4) - (1/4)*log2(1/4)) + (6/14)*((-4/6)*log2(4/6) -(1/3)*log2(1/3))
print(E_Temperature, '____ E_Temperature ____')

print()
IG_S_Temperature = Entrophy - E_Temperature
print(IG_S_Temperature, '____ IG_S_Temperature ____') # 0.0292225


## Entropy Windy ## 
print('______ Entropy Windy ________')
print()
E_Windy = (6/14)*((-3/6)*log2(3/6) -(3/6)*log2(3/6)) + (8/14)*((-6/8)*log2(6/8) - (2/8)*log2(2/8))
print(E_Windy, '____ E_Windy ____')
print()

print()
IG_S_Windy = Entrophy - E_Windy
print(IG_S_Windy, '____ IG_S_Windy ____') # 0.0481





## Entropy humidity ## 
print('______ Entropy Humidity ________')
print()
E_Humidity = (7/14)*((-6/7)*log2(6/7) -(1/7)*log2(1/7)) + (7/14)*((-3/7)*log2(3/7) - (4/7)*log2(4/7))
print(E_Humidity, '____ E_Humidity ____')
print()

print()
IG_S_Humidity = Entrophy - E_Humidity
print(IG_S_Humidity, '____ IG_S_Humidity ____') # 0.1518


# IG_S_Humidity 0.1518
# IG_S_Windy  0.0481
# IG_S_Temperature 0.0292225
# IG_S_Outlook -0.269


```
