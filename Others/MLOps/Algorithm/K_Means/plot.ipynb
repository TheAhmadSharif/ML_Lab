```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, cos, pi, linspace
import random
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])


plt.scatter(x, y, marker="*", edgecolors="green")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Test')
plt.grid()




for (key) in range(len(x)):
    label = "(" + str(x[key]) + "," + str(y[key]) + ")"
    plt.annotate(label,
                 (x[key],y[key]), 
                 textcoords="offset points", 
                 xytext=(0,10), 
                 ha='center')
plt.show()
```

```python
plt.scatter([1, 2, 3], [4, 5, 6])

plt.annotate("1", (1, 4))
plt.annotate("Point 1", (2, 5))
    

plt.show()
```

```python
# Circle


plt.plot(0,0, color = 'red', marker = 'o')

theta = np.linspace(0, 2*np.pi, 100)
r= 2
x = r*np.cos(theta)
y = r*np.sin(theta)
plt.plot(x,y)
plt.axis('equal')
plt.show()
```
