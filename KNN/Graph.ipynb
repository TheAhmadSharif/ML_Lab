```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

dataset = pd.read_csv("datasets/H_W_Gender.csv")
dataset = dataset[(dataset["Height"] < 200) & (dataset["Weight"] < 200)]

dataset.describe()
```

```python
# seaborn.lmplot() method is used to draw a scatter plot onto a FacetGrid.
# https://www.geeksforgeeks.org/python-seaborn-lmplot-method/
```

```python
sns.color_palette()
```

```python
sns.lmplot(data=dataset[:100], x="Height", y="Weight", hue="Sex", legend = "false", fit_reg = "false")
```

```python
sns.relplot(data=dataset[:500], x="Height", y="Weight", hue="Sex") # Rel plot : Relation Plot
```

```python
sns.catplot(data=dataset[:100], x="Height", y="Weight", hue="Sex") # cat plot : Categorical Plot
plt.xticks([110, 200]) 
```

```python
# create DataFrame 
df = pd.DataFrame({'var1': [25, 12, 15, 14, 19, 23, 25, 29], 
                   'var2': [5, 7, 7, 9, 12, 9, 9, 4]}) 
  
# create scatterplot 
sns.scatterplot(data=df, x='var1', y='var2') 
sns.scatterplot(data=[15, 8], x='var1', y='var2') 
  
# specify positions of ticks on x-axis and y-axis 
plt.xticks([15, 25]) 
plt.yticks([4, 12]) 
```
