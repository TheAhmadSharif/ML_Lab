```python
import nltk
from nltk.corpus import stopwords
import string

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')

text = """Moro terve. Meitsin Janne. Ja kyllä hyvät nettisivut maksaa. Ei siitä oikein mihinkään pääse ja miksi ne maksaa. 
No se ilmiselvä, lyhyt vastaus on se että niitten rakentamiseen kuluu tietenkin paljon aikaa ja siellä on monta asiaa mitkä pitää miettiä ja kun me kuitenkin halutaan tehdä asiat huolella me halutaan niinku laadukkaat sivut niin me ei voida oikeastaan oikaista pistää ne vaan tuossa katoin vähän paljon mä kulutan itse erilaisia projekteihin aikaan niin se alkaa niistä kehitysprojekteista niistä piiristä missä mä en edes niinku Mieti sitä suunnitelmaa. Mä vaan rakennan sen sivun jonkun muun suunnitelmasta 30 40 tunnista Ja sitten niinku päästään isompiin projekteihin oli yksi yksi projekti tuossa jossa Mulla meni 150 tuntia ja sitten jos sen 150 tuntia kertoo 90 perus Helsingissä Helsingissä menee yli kymppitonnin sitten se mutta nettisivua ei kannata katsoa pelkästään kulunavaimen ennemminkin sellaisena sijoituksena hyvin tärkeä myynnin työkaluja kun se on kohdillaan niin myyntikin sitten lisääntyy jos katsotaan että miten nopeasti ne sivut voi maksaa itsensä takaisin niin niin ehkä huomasitkin tän laskuri tuossa ylempänä sivulla voi olla et se on vähän muuttanut muotoansa mutta älä välitä siitä niin niin mä laitoin tänne vähän rohkeampia lukuja puhutaan nyt että Kymppitonni maksettaisiusta Miljoona vaihdettaisi jos se nettisivut lisää kymmenellä prosentilla sitä myyntiä kahdessa kuukaudessa maksanut itsensä takaisin eli tosi nopeasti jos ollaan vähän konservatiivisempia tai ihan vaan pessimistisiä et se parantaa vaan yhdellä prosentilla niin siltikin vuodessa se jo maksaa itsensä takaisin eli varsin niinku nopeasti yleensähän nettisivuja pidetään sellainen kolmisen vuotta No tietenkin jotkut pitää niitä hamaan tappiin asti mutta kolme kolme vuotta tota niin on ehkä sellainen Mitä että suosittelisin ne maksaa itsensä takaisin ja tämän takia ne kannattaa tehdä myös huolella jos jos me niinkun itse asioita puolella tai mietitään vaikka jotain henkilöä joka tekee halvemmalla ne sivut Joko se repii sen siitä omasta selkänahaastansa sen työn tai sitten oikaistaan jossain ei Joko mietitä Googlea ei mietitä asiakkaita Ei mietitä että mitä ne asiakkaat ylipäätänsä halua lukea sieltä nettisivulta tai ei ei mietitä välttämättä sitä visuaalista jo nämä asiat siellä varmaan Eikö mietitään että miten niinku se asiakas tai miten sitä asiakasta saadaan kannustettua siinä ostoprosessissa myyjien jostain """

words = nltk.word_tokenize(text.lower())

stop_words = set(stopwords.words('finnish'))

words_to_add = ["Ja", "ei", "kyllä", "paljon", "yksi", "jo", "mä"]

stop_words.update(words_to_add)

words = [word for word in words if word not in string.punctuation]

filtered_words = set(
    word.lower() for word in words 
    if word.lower() not in stop_words and not word.isnumeric() and word not in string.punctuation
)



for word in filtered_words:
  print(word, ' :: ')
```
