```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px

dataset = pd.read_csv("../datasets/H_W_Gender.csv")
dataset = dataset[(dataset["Height"] < 200) & (dataset["Weight"] < 200)]

dataset.describe()
```

```python
plt.plot([2,4], [4, 6], 'o:r')
plt.plot([2, 6, 9])
plt.hist([2, 4, 6])
plt.show() # Display all open figures.
```

```python
# seaborn.lmplot() method is used to draw a scatter plot onto a FacetGrid.
# https://www.geeksforgeeks.org/python-seaborn-lmplot-method/
```

```python
sns.color_palette()
```

```python
sns.lmplot(data=dataset[:100], x="Height", y="Weight", hue="Sex", legend = "false", fit_reg = "false")
```

```python
sns.relplot(data=dataset[:100], x="Height", y="Weight", hue="Sex") # Rel plot : Relation Plot
```

```python
fig = px.box(dataset[1:3000], x="Height", y="Weight", color="Sex")
fig.show()
```
