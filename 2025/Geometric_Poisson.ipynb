## Poisson Distribution ‚Äì Sports Examples

The **Poisson distribution** models the probability of a given number of events occurring in a fixed interval of time or space, assuming events occur independently and at a constant average rate.

---

### ‚öΩ 1. Football ‚Äì Goals per Match

**Problem:**  
A football team scores an average of 1.8 goals per match. What is the probability that the team scores **exactly 3 goals** in a match?

**Given:**

- $$ \lambda = 1.8 $$
- $$ x = 3 $$

**Manual Calculation:**

$$
P(X = 3) = \frac{e^{-\lambda} \cdot \lambda^x}{x!} = \frac{e^{-1.8} \cdot 1.8^3}{3!} = \frac{e^{-1.8} \cdot 5.832}{6}
$$

Calculate:

- \( e^{-1.8} \approx 0.1653 \)
- \( \frac{5.832}{6} = 0.972 \)

Therefore,

$$
P(X = 3) \approx 0.1653 \times 0.972 = 0.1606
$$

---

### üéæ 2. Tennis ‚Äì Aces per Set

**Problem:**  
A tennis player serves an average of 4 aces per set. What is the probability that the player serves **no aces** in a set?

**Given:**

- $$ \lambda = 4 $$
- $$ x = 0 $$

**Manual Calculation:**

$$
P(X = 0) = \frac{e^{-4} \cdot 4^0}{0!} = e^{-4} \approx 0.0183
$$

---

### üèÄ 3. Basketball ‚Äì Fouls per Game

**Problem:**  
A player commits an average of 2 fouls per game. What is the probability that the player commits **exactly 1 foul** in a game?

**Given:**

- $$ \lambda = 2 $$
- $$ x = 1 $$

**Manual Calculation:**

$$
P(X = 1) = \frac{e^{-2} \cdot 2^1}{1!} = 2 e^{-2} \approx 2 \times 0.1353 = 0.2707
$$

---

### ‚öΩ 4. Football ‚Äì Penalty Cards per Match

**Problem:**  
In a football match, penalty cards are given at an average rate of 0.2 per match. What is the probability that **no penalty cards** are given in a match?

**Given:**

- $$ \lambda = 0.2 $$
- $$ x = 0 $$

**Manual Calculation:**

$$
P(X = 0) = e^{-0.2} \approx 0.8187
$$

---

### ‚öæ 5. Baseball ‚Äì Home Runs per Game

**Problem:**  
A baseball player hits an average of 0.5 home runs per game. What is the probability that the player hits **exactly 2 home runs** in a game?

**Given:**

- $$ \lambda = 0.5 $$
- $$ x = 2 $$

**Manual Calculation:**

$$
P(X = 2) = \frac{e^{-0.5} \cdot 0.5^2}{2!} = \frac{e^{-0.5} \cdot 0.25}{2} = 0.125 \times e^{-0.5}
$$

Calculate:

- \( e^{-0.5} \approx 0.6065 \)

Therefore,

$$
P(X = 2) \approx 0.125 \times 0.6065 = 0.0758
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Define scenarios
examples = [
    {"label": "Football: Goals per Match", "lambda": 1.8, "x": 3},
    {"label": "Tennis: Aces per Set", "lambda": 4, "x": 0},
    {"label": "Basketball: Fouls per Game", "lambda": 2, "x": 1},
    {"label": "Football: Penalties per Match", "lambda": 0.2, "x": 0},
    {"label": "Baseball: Home Runs per Game", "lambda": 0.5, "x": 2},
]

# Calculate probabilities and plot
fig, axs = plt.subplots(3, 2, figsize=(12, 12))
axs = axs.flatten()

for i, example in enumerate(examples):
    lam = example["lambda"]
    x_val = example["x"]
    label = example["label"]

    # Values to plot
    x = np.arange(0, max(6, x_val + 3))
    pmf = poisson.pmf(x, lam)

    axs[i].bar(x, pmf, color='skyblue', edgecolor='black')
    axs[i].set_title(f"{label} (Œª={lam})")
    axs[i].set_xlabel("x")
    axs[i].set_ylabel("P(X = x)")
    axs[i].axvline(x=x_val, color='red', linestyle='--', label=f"P(X={x_val}) ‚âà {poisson.pmf(x_val, lam):.4f}")
    axs[i].legend()

# Hide the last subplot if odd number of plots
if len(examples) % 2 != 0:
    axs[-1].axis('off')

plt.tight_layout()
plt.show()
```

# üìä Geometric vs Poisson Distributions

## 1. Comparison: Geometric vs Poisson Distributions

| Feature               | Geometric Distribution                         | Poisson Distribution                          |
|----------------------|-------------------------------------------------|-----------------------------------------------|
| **Type**             | Discrete                                        | Discrete                                      |
| **Describes**        | Number of trials until the first success        | Number of events in a fixed interval          |
| **Typical Use Case** | First success in Bernoulli trials               | Count of events over time or space            |
| **Key Parameter(s)** | \( p \): Probability of success on each trial   | \( \lambda \): Average rate of events         |
| **Memoryless**       | ‚úÖ Yes                                           | ‚ùå No                                          |
| **Support**          | \( x \in \{1, 2, 3, \ldots\} \)                 | \( x \in \{0, 1, 2, \ldots\} \)               |

### üìå PMF Formulas

**Geometric Distribution**:

$$
P(X = x) = (1 - p)^{x - 1} \cdot p, \quad x = 1, 2, 3, \ldots
$$

**Poisson Distribution**:

$$
P(X = x) = \frac{e^{-\lambda} \cdot \lambda^x}{x!}, \quad x = 0, 1, 2, \ldots
$$

---

## 2. Examples

### üéØ Geometric Distribution Examples

#### ‚úÖ Example 1: Coin Toss

- **Scenario**: Tossing a fair coin until the first heads.
- **Probability of success**: \( p = 0.5 \)
- **Find**: Probability that first heads appears on the 3rd toss.

$$
P(X = 3) = (1 - 0.5)^2 \cdot 0.5 = 0.125
$$

---

#### ‚úÖ Example 2: Sales Call Success

- **Scenario**: Salesperson has 10% chance to make a sale per call.
- **Probability of success**: \( p = 0.1 \)
- **Find**: Probability that the first sale happens on the 5th call.

$$
P(X = 5) = (1 - 0.1)^4 \cdot 0.1 = 0.6561 \cdot 0.1 = 0.06561
$$

---

### üí• Poisson Distribution Examples

#### ‚úÖ Example 1: Call Center

- **Scenario**: Call center receives 5 calls per minute on average.
- **Rate**: \( \lambda = 5 \)
- **Find**: Probability of getting exactly 3 calls in a minute.

$$
P(X = 3) = \frac{e^{-5} \cdot 5^3}{3!} \approx 0.1404
$$

---

#### ‚úÖ Example 2: Football Goals

- **Scenario**: A team scores 2.5 goals per match on average.
- **Rate**: \( \lambda = 2.5 \)
- **Find**: Probability they score exactly 4 goals.

$$
P(X = 4) = \frac{e^{-2.5} \cdot 2.5^4}{4!} \approx 0.133
$$


# üìä Hypergeometric vs. Binomial Distribution

## üß† Conceptual Comparison

| Feature                     | Hypergeometric Distribution                                         | Binomial Distribution                                         |
|----------------------------|---------------------------------------------------------------------|---------------------------------------------------------------|
| **Scenario**               | Sampling **without replacement**                                   | Sampling **with replacement** or independent trials           |
| **Population Size (N)**    | Finite and fixed                                                    | Can be considered infinite or not explicitly defined          |
| **Random Variable (X)**    | Number of successes in a sample of size \( n \)                    | Number of successes in \( n \) independent trials             |
| **Dependence**             | Dependent trials (because of no replacement)                        | Independent trials                                            |
| **PMF**                    | \( P(X = x) = \frac{{\binom{K}{x} \binom{N - K}{n - x}}}{\binom{N}{n}} \) | \( P(X = x) = \binom{n}{x} p^x (1 - p)^{n - x} \)             |
| **Mean**                   | \( \mu = n \cdot \frac{K}{N} \)                                     | \( \mu = n \cdot p \)                                         |
| **Variance**               | \( \sigma^2 = n \cdot \frac{K}{N} \cdot \frac{N - K}{N} \cdot \frac{N - n}{N - 1} \) | \( \sigma^2 = n \cdot p \cdot (1 - p) \)                     |

---

## üéØ Example: Hypergeometric Distribution

**Problem**: A box contains 10 light bulbs, 4 of which are defective. If 3 bulbs are selected **without replacement**, what is the probability that exactly 2 are defective?

### Parameters:
- \( N = 10 \), total items
- \( K = 4 \), number of defective items
- \( n = 3 \), number selected
- \( x = 2 \), number of defectives desired

### PMF:
\[
P(X = 2) = \frac{\binom{4}{2} \cdot \binom{6}{1}}{\binom{10}{3}} = \frac{6 \cdot 6}{120} = \frac{36}{120} = 0.3
\]

---

## üéØ Example: Binomial Distribution

**Problem**: A machine produces bulbs with a 0.4 probability of being defective. If 3 bulbs are selected **with replacement**, what is the probability that exactly 2 are defective?

### Parameters:
- \( n = 3 \), number of trials
- \( p = 0.4 \), probability of success (defective)
- \( x = 2 \)

### PMF:
\[
P(X = 2) = \binom{3}{2} \cdot 0.4^2 \cdot 0.6^1 = 3 \cdot 0.16 \cdot 0.6 = 0.288
\]

---

## ‚úÖ Summary

- Use **Hypergeometric** when sampling **without replacement** from a finite population.
- Use **Binomial** when trials are **independent**, usually with replacement or infinite population.
