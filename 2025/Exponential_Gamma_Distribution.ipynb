## Exponential Distribution and the Poisson Process

We now study a **continuous distribution** that arises naturally from the **Poisson process**.

### ðŸ” From Poisson (Discrete) to Exponential (Continuous)

- In a **Poisson process**, we often count the **number of events** in a fixed interval of time. That count is a **discrete random variable** with a **Poisson distribution**.
- But we can also ask a different question:
  
  > "How long until the **first** event occurs?"

- This **waiting time** is a **continuous** random variable.

---

### ðŸ“ Defining the Exponential Distribution

Let:

- $$W$$ = the waiting time until the **first occurrence**
- $$\lambda$$ = the **average rate** (mean number of events per unit time)

Then $$W$$ follows an **Exponential distribution** with parameter $$\lambda$$:

$$
W \sim \text{Exponential}(\lambda)
$$

---

### ðŸ“Š Probability Density Function (PDF)

The PDF of the exponential distribution is:

$$
f(w) = \lambda e^{-\lambda w}, \quad w > 0
$$

---

### ðŸ“ˆ Cumulative Distribution Function (CDF)

The CDF (probability that the first event occurs within time $$w$$) is:

$$
F(w) = P(W \le w) = 1 - e^{-\lambda w}, \quad w > 0
$$

---

### ðŸ§  Example

Suppose:
- On average, a phone rings **3 times per hour**, so $$\lambda = 3$$.
- What is the probability that the first ring comes **within 10 minutes**?

10 minutes = $$\frac{1}{6}$$ hour.

Use the CDF:

$$
F\left(\frac{1}{6}\right) = 1 - e^{-3 \cdot \frac{1}{6}} = 1 - e^{-0.5} \approx 1 - 0.6065 = 0.3935
$$

ðŸ‘‰ There is approximately a **39.35% chance** the phone rings within 10 minutes.

---

### ðŸ“Œ Summary

- The **Poisson distribution** counts events in an interval (discrete).
- The **Exponential distribution** models **waiting time until the first event** (continuous).
- It has the important **memoryless property**:

$$
P(W > t + s \mid W > s) = P(W > t)
$$

This makes it especially useful in modeling time between independent random events.

---


## ðŸ“š Examples of the Exponential Distribution

The Exponential distribution models the **time between events** in a **Poisson process**. It is used when:

- Events happen **independently**
- Events occur at a **constant average rate**
- We are interested in the **waiting time until the next event**

---

### ðŸ”” Example 1: Phone Calls at a Call Center

- Calls arrive at a call center at an average rate of $$\lambda = 10$$ calls per hour.
- Let $$W$$ be the **waiting time (in hours)** until the next call.
- Then:

$$
W \sim \text{Exponential}(10)
$$

---

### ðŸ’¡ Example 2: Light Bulb Lifespan

- A type of light bulb has an **average life** of 1000 hours.
- That means $$\lambda = \frac{1}{1000}$$ (failures per hour).
- Let $$T$$ be the **time until failure** for one bulb.
- Then:

$$
T \sim \text{Exponential}\left(\frac{1}{1000}\right)
$$

---

### ðŸ’‰ Example 3: Time Between Patient Arrivals

- Patients arrive at a hospital emergency room at a rate of 5 per hour.
- Let $$X$$ be the **time (in hours)** between two arrivals.
- Then:

$$
X \sim \text{Exponential}(5)
$$

---

### â›½ Example 4: Time Until a Gas Station Gets a Customer

- On average, one customer arrives every 12 minutes.
- That means $$\lambda = \frac{1}{12}$$ (customers per minute).
- Let $$Y$$ be the **waiting time** (in minutes) until the next customer.
- Then:

$$
Y \sim \text{Exponential}\left(\frac{1}{12}\right)
$$

---

### ðŸ§ª Example 5: Radioactive Decay

- In physics, the **time between decay events** of a radioactive atom follows an exponential distribution.
- If the average decay rate is $$\lambda$$ decays per second, then:

$$
\text{Time until next decay} \sim \text{Exponential}(\lambda)
$$

---

These examples illustrate how the exponential distribution is used to model **time-to-next-event** scenarios in fields like engineering, healthcare, physics, and telecommunications.

## ðŸ§® Exponential Distribution Example: Customer Arrival

Customers arrive at a shop following an **approximate Poisson process** with a mean rate of **20 customers per hour**.

---

### ðŸ“Œ Given

- Arrival rate: $$20 \text{ per hour} = \frac{1}{3} \text{ per minute}$$
- Let $$X$$ = waiting time (in minutes) until the **first customer arrives**
- So the rate parameter is:
  
  $$
  \lambda = \frac{1}{3}, \quad \theta = \frac{1}{\lambda} = 3
  $$

---

### ðŸ“ˆ PDF of the Exponential Distribution

The probability density function is:

$$
f(x) = \frac{1}{3} e^{-\frac{1}{3}x}, \quad \text{for } x \ge 0
$$

---

### â“ What is the probability that the shopkeeper waits **more than 5 minutes**?

We compute:

$$
P(X > 5) = \int_5^\infty \frac{1}{3} e^{-\frac{1}{3}x} \, dx
$$

This simplifies to:

$$
P(X > 5) = e^{-5/3} = 0.1889
$$

ðŸ§¾ **Interpretation**: There's about an **18.89% chance** that the shopkeeper waits **more than 5 minutes** for the first customer.

---

### ðŸ”¹ Median Waiting Time

The median time \( m \) satisfies:

$$
P(X \le m) = 0.5
$$

Using the exponential CDF:

$$
1 - e^{-\frac{1}{3}m} = 0.5 \quad \Rightarrow \quad e^{-\frac{1}{3}m} = 0.5
$$

Take the natural logarithm:

$$
-\frac{1}{3}m = \ln(0.5) \quad \Rightarrow \quad m = -3 \ln(0.5) = 2.0794
$$

ðŸ“Œ **Median waiting time** = **2.0794 minutes**

# Gamma Distribution

## ðŸ“š Gamma Distribution

The **Gamma distribution** is a **continuous probability distribution** used to model the **waiting time** until the **k-th event** in a Poisson process.

---

### ðŸ§  Intuition

- **Exponential distribution** models the time until the **first** event.
- **Gamma distribution** models the time until the **k-th** event.

---

### ðŸ”¢ Parameters

Let $$X \sim \text{Gamma}(\alpha, \lambda)$$ where:

- $$\alpha > 0$$ is the **shape** parameter (number of events)
- $$\lambda > 0$$ is the **rate** parameter (events per unit time)
- Sometimes we use $$\theta = \frac{1}{\lambda}$$ as the **scale** parameter

---

### ðŸ“ Probability Density Function (PDF)

$$
f(x) = \frac{\lambda^\alpha x^{\alpha - 1} e^{-\lambda x}}{\Gamma(\alpha)}, \quad x > 0
$$

Where:

- $$\Gamma(\alpha) = \int_0^\infty t^{\alpha - 1} e^{-t} dt$$ is the **gamma function**, which generalizes factorial:
  
  $$
  \Gamma(n) = (n - 1)! \quad \text{for integer } n
  $$

---

### ðŸ“Š Mean and Variance

- **Mean**:
  $$
  \mathbb{E}[X] = \frac{\alpha}{\lambda}
  $$
- **Variance**:
  $$
  \text{Var}(X) = \frac{\alpha}{\lambda^2}
  $$

---

### ðŸ”¹ Special Cases

- When $$\alpha = 1$$, the gamma distribution becomes the **exponential distribution**:
  
  $$
  \text{Gamma}(1, \lambda) = \text{Exponential}(\lambda)
  $$

- When $$\alpha$$ is a **positive integer**, it's also called the **Erlang distribution**
