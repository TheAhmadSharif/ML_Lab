## ðŸ§  What Makes These Distributions Different?

Most common distributions (like **binomial**, **normal**, **exponential**) describe **real-world events**.

But the **t-distribution**, **chi-squared distribution**, and **F-distribution** are different:

They donâ€™t describe something that happens **in nature** directly.

Instead, they come from **doing calculations on sample data** when performing **statistical inference**.

---

## ðŸ”¸ They Don't Model Real-World Events Directly

| Distribution        | Models Real-World Events?                  |
|---------------------|--------------------------------------------|
| **Binomial**         | âœ… Yes â€“ number of successes in coin flips |
| **Exponential**      | âœ… Yes â€“ time until something happens       |
| **Normal**           | âœ… Yes â€“ measurements like height or weight|
| **t-distribution**   | âŒ No â€“ used for inference on small samples|
| **Chi-square**       | âŒ No â€“ used to test fit/independence      |
| **F-distribution**   | âŒ No â€“ used to compare variances          |

---

## ðŸ”¬ They Arise From Sample Calculations

### ðŸ“ **t-distribution**

Used when:
- You're estimating a population **mean**
- You have a **small sample size**
- Population standard deviation is **unknown**

**Example:**

> "Is the average height of a class different from 160 cm?"

If you don't know the population standard deviation, use the **t-distribution**.

---

### ðŸ§® **Chi-squared distribution**

Used when:
- You're checking how well **observed data** match **expected values**
- You're testing **independence** in a contingency table
- You're estimating **population variance**

Defined as:

$$
\chi^2 = \sum_{i=1}^n Z_i^2 \quad \text{where } Z_i \sim \mathcal{N}(0,1)
$$

**Example:**

> "Are red, green, and blue M&Ms equally likely?"

Use the **chi-squared test** to see if observed counts match expectations.

---

### âš–ï¸ **F-distribution**

Used when:
- You want to compare **two variances**
- You are doing **ANOVA** to compare group means

Defined as:

$$
F = \frac{U_1 / d_1}{U_2 / d_2}, \quad \text{where } U_1, U_2 \sim \chi^2 \text{ and independent}
$$

**Example:**

> "Is the variation in test scores between two schools significantly different?"

Use the **F-distribution** to compare their sample variances.

---

## âœ… Summary Table

| Distribution      | What It Helps With                                |
|------------------|----------------------------------------------------|
| **t**            | Compare sample mean with population mean (small $$n$$) |
| **Chi-square**   | Test fit to expected values / test independence     |
| **F**            | Compare how **spread out** (variances) two groups are |

---

## ðŸŽ¯ Big Picture

Think of it like this:

> Natural distributions = What nature gives you (e.g., coin flips, wait times, measurements)  
>  
> t, chi-squared, F = **Math tools** we use on samples to draw conclusions about populations

They are **derived distributions** â€“ useful for **inference**, not directly for modeling nature.

---


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, chi2, f

# Create x-values for each distribution
x_t = np.linspace(-5, 5, 500)
x_chi2 = np.linspace(0, 20, 500)
x_f = np.linspace(0.01, 5, 500)

# Parameters
dfs_t = [1, 5, 10, 30]
dfs_chi2 = [1, 2, 5, 10]
dfs_f = [(5, 2), (5, 5), (10, 10), (20, 20)]

# Plot t-distribution
plt.figure(figsize=(16, 4))
plt.subplot(1, 3, 1)
for df in dfs_t:
    plt.plot(x_t, t.pdf(x_t, df), label=f'df={df}')
plt.title('t-Distribution')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.grid(True)

# Plot Chi-squared distribution
plt.subplot(1, 3, 2)
for df in dfs_chi2:
    plt.plot(x_chi2, chi2.pdf(x_chi2, df), label=f'df={df}')
plt.title('Chi-squared Distribution')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.grid(True)

# Plot F-distribution
plt.subplot(1, 3, 3)
for d1, d2 in dfs_f:
    plt.plot(x_f, f.pdf(x_f, d1, d2), label=f'df1={d1}, df2={d2}')
plt.title('F-Distribution')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
