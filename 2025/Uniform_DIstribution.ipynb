## Uniform Distribution: Theory and Examples

The **Uniform distribution** is a continuous probability distribution where every value in an interval \([a, b]\) is equally likely.

### Definition:

A continuous random variable \(X\) has a **Uniform distribution** on \([a, b]\), denoted \(X \sim U(a, b)\), if its probability density function (pdf) is constant on \([a, b]\) and zero elsewhere:

$$
f(x) = \frac{1}{b - a}, \quad a \leq x \leq b
$$

The total area under the pdf curve is 1, satisfying:

$$
\int_a^b f(x) \, dx = 1
$$

### Properties:

- **Mean (expected value):**

$$
\mu = E(X) = \frac{a + b}{2}
$$

- **Variance:**

$$
\sigma^2 = Var(X) = \frac{(b - a)^2}{12}
$$

---

### Examples:

1. **Waiting time between 0 and 30 minutes:**

If your friend arrives uniformly between 0 and 30 minutes,

$$
X \sim U(0, 30)
$$

then the pdf is

$$
f(x) = \frac{1}{30}, \quad 0 \leq x \leq 30
$$

and the mean arrival time is

$$
\mu = \frac{0 + 30}{2} = 15 \text{ minutes}
$$

2. **Rolling a fair die (modeled continuously):**

If you imagine a fair die roll modeled as a uniform distribution over the numbers 1 to 6,

$$
X \sim U(1, 6)
$$

then the pdf is

$$
f(x) = \frac{1}{5}, \quad 1 \leq x \leq 6
$$

and the expected value is

$$
\mu = \frac{1 + 6}{2} = 3.5
$$

---

### Summary:

- Uniform distribution models situations where all outcomes in an interval are equally likely.
- The pdf is flat (rectangular shape).
- Useful for modeling random arrival times, random measurements, or fair selections within a range.
